```python
import pandas as pd
from bokeh.io import output_notebook
output_notebook()
bokeh_width, bokeh_height = 1000,800
```

```python
from scipy.io import loadmat
data = loadmat('../matlab/results_real/20210602_i5_Alcorcon1.mat')
```

```python
import os 
api_key = os.environ['GOOGLE_API_KEY']
```

```python
lat, lon = 40.42040236306583, -3.664199622028989
```

```python
nlls = data['nlls']
lls = data['lls']
```

```python
from bokeh.io import show
from bokeh.plotting import gmap
from bokeh.models import GMapOptions

def plot(lat, lng, data, zoom=10, map_type='roadmap', color='red'):
    gmap_options = GMapOptions(lat=lat, lng=lng, 
                               map_type=map_type, zoom=zoom)
    p = gmap(api_key, gmap_options, title='Madrid', 
             width=bokeh_width, height=bokeh_height)
    
    center = p.circle([lng], [lat], size=10, color='magenta')
    for solution in data:
        point = p.circle([solution[1]],[solution[0]], size=10, alpha=1, color=color)
    
    show(p)
    return p
```

```python
p = plot(lat, lon, nlls, zoom=15, map_type='satellite')
```

```python
p_lls = plot(lat, lon, lls, zoom=15, map_type='satellite', color='blue')
```
